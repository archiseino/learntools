### Types of Charts in Seaborn
Visualize trends over time

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set a theme for Seaborn
sns.set_theme(style="whitegrid")

# Create a custom dataset
data = {
    "Category": np.random.choice(["A", "B", "C"], size=100),  # Categorical data
    "Subcategory": np.random.choice(["X", "Y", "Z"], size=100),  # Another categorical data
    "Value1": np.random.randint(10, 100, size=100),  # Numerical data
    "Value2": np.random.uniform(1.0, 10.0, size=100),  # Numerical data
    "Boolean": np.random.choice([True, False], size=100)  # Boolean data
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataset
print("Dataset Preview:")
print(df.head())

# Visualizations
# 1. Count plot of 'Category'
plt.figure(figsize=(8, 5))
sns.countplot(x="Category", data=df, palette="Set2")
plt.title("Count of Categories")
plt.show()

# 2. Boxplot of 'Value1' grouped by 'Category'
plt.figure(figsize=(8, 5))
sns.boxplot(x="Category", y="Value1", data=df, palette="Set3")
plt.title("Boxplot of Value1 by Category")
plt.show()

# 3. Scatter plot of 'Value1' vs 'Value2' with 'Category' as hue
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Value1", y="Value2", hue="Category", style="Subcategory", data=df, palette="viridis")
plt.title("Scatter Plot of Value1 vs Value2")
plt.show()

# 4. Pairplot of the numerical features colored by 'Category'
sns.pairplot(df, vars=["Value1", "Value2"], hue="Category", palette="coolwarm")
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()

# 5. Heatmap of correlations among numerical features
plt.figure(figsize=(8, 5))
correlation_matrix = df[["Value1", "Value2"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# 6. Distribution plot of 'Value1'
plt.figure(figsize=(8, 5))
sns.histplot(df["Value1"], kde=True, color="blue", bins=15)
plt.title("Distribution of Value1")
plt.xlabel("Value1")
plt.ylabel("Frequency")
plt.show()

# 7. KDE plot of 'Value2' grouped by 'Category'
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x="Value2", hue="Category", fill=True, alpha=0.5, palette="Set1")
plt.title("KDE Plot of Value2 by Category")
plt.xlabel("Value2")
plt.ylabel("Density")
plt.show()

# 8. Line plot showing cumulative sum of 'Value1' by 'Category'
cumsum_df = df.groupby("Category")["Value1"].cumsum()
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x=df.index, y=cumsum_df, hue="Category", palette="cool")
plt.title("Cumulative Sum of Value1 by Category")
plt.xlabel("Index")
plt.ylabel("Cumulative Sum of Value1")
plt.show()
```
