## Feature Engineering
Steps and principles of creating a better features

### Goal of Feature Engineering
The goal of feature engineering is simply to make your data better suited to the problem at hand.

### Guide to Feature Engineering
For a feature to be useful, it must have a relationship to the `target` your model is able to learn.

For instance, Linear Model are only able to learn the linar relationship. So, your goal is to transform features to make their relationship to the target linear.

Let's look a case on the `Price` prediction based on the `Length` of one side.      



Since this was only length, we can square those to get an Area, Squaring those will resulted in a fit model.
